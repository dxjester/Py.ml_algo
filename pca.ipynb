# Principle Component Analysis

Import the necessary modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time as t
```

Import the breast cancer data set

```python
from sklearn.datasets import load_breast_cancer
```

```python
raw = load_breast_cancer()
raw_data = raw.data
raw_data
```

```python
raw_data.shape
```

```python
data_labels = raw.target
data_labels
```

```python
data_labels.shape
```

```python
labels = np.reshape(data_labels, (569,1))
labels
```

```python
data_final = np.concatenate([raw_data, labels], axis=1)
```

```python
data_final.shape
```

```python
breast_data = pd.DataFrame(data_final)
breast_data.head(10)
```

Import the features

```python
features = raw.feature_names
features
```

```python
features_labels = np.append(features, 'label')
```

```python
breast_data.columns = features_labels
breast_data.head(5)
```
